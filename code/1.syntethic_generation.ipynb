```python
# Parameters
input_dir = 'data'

countries_iso3_filter = ['RUS','URK']
year = '2017'

import os
from semantic_relatedness import synthetic

working_dir = input_dir
os.chdir(working_dir)
```

```python
import pandas

kb = pandas.read_csv(os.path.join(input_dir, 'RUS_UKN_Coverage_2017.csv'), index_col=None, header=0)

relations = kb['r'].unique().tolist()
entities = pandas.concat([kb['h'], kb['t'] ]).unique().tolist()
```

```python
print(kb.values.tolist())
```

```python
kg = kb.values.tolist()
synth = list()
for r in relations:
  r_synth = synthetic(r)
  for i in entities:
    for j in entities:
      if (i != j) and (kg.count([i,r,j])==0) :
        synth.append([i,r_synth,j])

triples = pandas.DataFrame(synth, columns=['h', 'r', 't'])
triples.drop_duplicates(inplace=True)
triples.sort_values(['r','r'], inplace=True)
print('Triples:', len(triples))
triples.to_csv(os.path.join(input_dir, 'RUS_UKN_Coverage_2017_synth.csv'), index=False)
```
