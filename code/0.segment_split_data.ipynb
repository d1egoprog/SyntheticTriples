```python
import pandas

countries = pandas.read_csv('Countries_V202211.csv', low_memory=False, index_col=None)
countries['country'] = countries['country'].str.replace(' ','').replace(',','')
countryDict = dict(zip(countries['iso3'],countries['country']))

coverage = pandas.read_csv('Label_rta_coverage_V202211.csv', low_memory=False, index_col=None)
coverage['rta_coverage_lbl'] = coverage['rta_coverage_lbl'].str.replace(' & ','_')
coverageDict = dict(zip(coverage['rta_coverage'],coverage['rta_coverage_lbl']))

types = pandas.read_csv('Label_rta_type_V202211.csv', low_memory=False, index_col=None)
types['rta_type_lbl'] = types['rta_type_lbl'].str.replace(' & ','_')
typesDict = dict(zip(types['rta_type'],types['rta_type_lbl']))
```

```python
df_war = pandas.read_csv('Gravity_{}.csv'.format(year))

df_war['country_full_o'] = df_war['iso3_o'].copy().map(countryDict).values
df_war['country_full_d'] = df_war['iso3_d'].copy().map(countryDict).values

df_war['rta_coverage'] = df_war['rta_coverage'].copy().map(coverageDict).values
df_war['rta_type'] = df_war['rta_type'].copy().map(typesDict).values

df_war = df_war.loc[(df_war['country_exists_o']==1)]
df_war = df_war.loc[(df_war['country_exists_d']==1)]
df_war = df_war.loc[(df_war['iso3_o']!=df_war['iso3_d'])]

df_war_o = df_war[df_war['iso3_o'].isin(countries_iso3_filter)]
df_war_d = df_war[df_war['iso3_d'].isin(countries_iso3_filter)]
df_war_def = pandas.concat([df_war_o, df_war_d])

selected_columns = ['iso3_o', 'iso3_d', 'country_full_o', 'country_full_d', 'rta_coverage', 'rta_type' ]
df_war_def = df_war[selected_columns]

df_war_def = df_war_def.fillna(0)

df_war_def
```
