```python
import os
# Parameters
input_dir = 'data'
year = '2017'
working_dir = input_dir
os.chdir(working_dir)
```

```python
import pickle as pk
import numpy as np
import pandas

from ampligraph.datasets import load_from_csv
from ampligraph.evaluation import train_test_split_no_unseen
```

```python
# Load the dataset
X = load_from_csv('.', 'RUS_UKN_Coverage_2017.csv', sep=',', header=0)[:, :3]
print(X.shape)

num_val = int(len(X) * 0.20) #0.020)   # Validation is 18% dataset
num_test = int(len(X) * 0.20) #0.020)  # Test is 10% dataset

X_train_valid, X_test = train_test_split_no_unseen(X, test_size=num_test, allow_duplication=True)
X_train, X_valid = train_test_split_no_unseen(X_train_valid, test_size=num_val, allow_duplication=True)

print('Train set size: ', X_train.shape)
print('Test set size: ', X_test.shape)
print('Validation set size: ', X_valid.shape)

with open('RUS_UKN_Coverage_2017_splited.data', 'wb') as f:
    pk.dump([X_train, X_valid, X_test], f)
```
